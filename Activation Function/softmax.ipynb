```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi Softmax
def softmax(z):
    exp_z = np.exp(z - np.max(z))  # untuk stabilitas numerik
    return exp_z / exp_z.sum(axis=0)

# Nilai input z
z = np.array([1.0, 2.0, 3.0, 4.0, 1.0])
# Nilai Softmax untuk setiap z
softmax_values = softmax(z)

# Plot fungsi Softmax
plt.figure(figsize=(8, 6))
plt.bar(range(len(z)), softmax_values, color='b', alpha=0.7)
plt.xlabel('Class')
plt.ylabel('Probability')
plt.title('Fungsi Softmax')
plt.grid(True)
plt.show()
```
