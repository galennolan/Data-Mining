ReLU (Rectified Linear Unit) adalah fungsi aktivasi keren yang bikin nilai negatif jadi nol dan biarkan nilai positif tetap ada. Fungsi ini bantu neural network belajar lebih cepat dan mengatasi masalah vanishing gradient.

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi ReLU
def relu(z):
    return np.maximum(0, z)

# Rentang nilai z
z = np.linspace(-10, 10, 100) # Menghasilkan 100 titik antara -10 dan 10
# Nilai ReLU untuk setiap z
relu_values = relu(z)

# Plot fungsi ReLU
plt.plot(z, relu_values)
plt.xlabel('z')
plt.ylabel('ReLU(z)')
plt.title('Fungsi ReLU')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi ReLU
def relu(z):
    return np.maximum(0, z)

# Nilai input z
z = np.array([-10, -5, -2, -1, 0, 1, 2, 5, 10])
# Nilai ReLU untuk setiap z
relu_values = relu(z)

# Plot fungsi ReLU
plt.figure(figsize=(8, 6))
plt.plot(z, relu_values, marker='o', linestyle='-', color='b')
plt.xlabel('Input (z)')
plt.ylabel('Output (ReLU(z))')
plt.title('Fungsi ReLU (Rectified Linear Unit)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(z):
    return np.maximum(0, z)

z = np.linspace(-5, 5, 100)
relu_vals = relu(z)

plt.figure(figsize=(8, 4))
plt.plot(z, relu_vals, label='ReLU(z)')
plt.xlabel('z (Input)')
plt.ylabel('ReLU(z) (Output)')
plt.title('Fungsi ReLU')
plt.grid(True)
plt.legend()
plt.show()
```
