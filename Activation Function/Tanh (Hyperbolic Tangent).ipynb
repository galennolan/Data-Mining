Fungsi Tanh (Hyperbolic Tangent) adalah fungsi aktivasi yang mengonversi nilai input menjadi rentang antara -1 dan 1, yang bisa lebih baik dari sigmoid dalam beberapa kasus karena outputnya terpusat di sekitar nol. Fungsi Tanh membantu dalam membuat data terpusat di sekitar nol, yang bisa membuat jaringan saraf lebih stabil dan konvergen lebih cepat.

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi Tanh
def tanh(z):
    return np.tanh(z)

# Nilai input z
z = np.array([-10, -5, -2, -1, 0, 1, 2, 5, 10])
# Nilai Tanh untuk setiap z
tanh_values = tanh(z)

# Plot fungsi Tanh
plt.plot(z, tanh_values, marker='o')
plt.xlabel('z')
plt.ylabel('tanh(z)')
plt.title('Fungsi Tanh')
plt.grid(True)
plt.show()
```
