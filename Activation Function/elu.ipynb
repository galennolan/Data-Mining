```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi ELU
def elu(z, alpha=1.0):
    return np.where(z > 0, z, alpha * (np.exp(z) - 1))

# Nilai input z
z = np.array([-10, -5, -2, -1, 0, 1, 2, 5, 10])
# Nilai ELU untuk setiap z
elu_values = elu(z)

# Plot fungsi ELU
plt.figure(figsize=(8, 6))
plt.plot(z, elu_values, marker='o', linestyle='-', color='b')
plt.xlabel('z')
plt.ylabel('ELU(z)')
plt.title('Fungsi ELU (Exponential Linear Unit)')
plt.grid(True)
plt.show()
```
