Leaky ReLU adalah modifikasi dari fungsi ReLU yang memungkinkan beberapa nilai negatif lolos dengan kemiringan kecil. Ini membantu mengatasi masalah "dead neurons" dalam jaringan saraf di mana neuron dapat menjadi tidak aktif jika mereka hanya menerima input negatif.

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi Leaky ReLU
def leaky_relu(z, alpha=0.01):
    return np.where(z > 0, z, alpha * z)

# Nilai input z
z = np.array([-10, -5, -2, -1, 0, 1, 2, 5, 10])
# Nilai Leaky ReLU untuk setiap z
leaky_relu_values = leaky_relu(z)

# Plot fungsi Leaky ReLU
plt.plot(z, leaky_relu_values, marker='o')
plt.xlabel('z')
plt.ylabel('Leaky ReLU(z)')
plt.title('Fungsi Leaky ReLU')
plt.grid(True)
plt.show()
```
