```python
import matplotlib.pyplot as plt
import numpy as np

# Data fiktif
np.random.seed(0)
total_sales = np.random.randint(1000, 10000, 100)
price = np.random.randint(10, 100, 100)

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(price, total_sales, alpha=0.5)
plt.title('Hubungan Total Penjualan dengan Harga Produk')
plt.xlabel('Harga Produk')
plt.ylabel('Total Penjualan')
plt.show()
```

```python
import pandas as pd

# Data fiktif
regions = ['Jakarta', 'Bandung', 'Surabaya', 'Yogyakarta', 'Semarang']
sales_data = {'Region': [], 'Total Sales': []}
for region in regions:
    sales_data['Region'].append(region)
    sales_data['Total Sales'].append(np.random.randint(10000, 50000))

df_sales = pd.DataFrame(sales_data)

# Bar chart
plt.figure(figsize=(10, 6))
plt.bar(df_sales['Region'], df_sales['Total Sales'], color='skyblue')
plt.title('Total Penjualan per Wilayah')
plt.xlabel('Wilayah')
plt.ylabel('Total Penjualan')
plt.show()
```
