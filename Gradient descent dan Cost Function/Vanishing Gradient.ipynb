```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Gradien dari Fungsi Sigmoid
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Rentang nilai x
x = np.linspace(-10, 10, 100)

# Hitung nilai sigmoid dan gradien sigmoid
sigmoid_values = sigmoid(x)
gradient_values = sigmoid_derivative(x)

# Plot fungsi sigmoid
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(x, sigmoid_values, label='Sigmoid')
plt.title('Fungsi Sigmoid')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.grid(True)
plt.legend()

# Plot gradien dari sigmoid
plt.subplot(1, 2, 2)
plt.plot(x, gradient_values, label='Gradien Sigmoid', color='r')
plt.title('Gradien dari Fungsi Sigmoid')
plt.xlabel('x')
plt.ylabel('Gradien')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
