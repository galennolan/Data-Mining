```python
import matplotlib.pyplot as plt
import numpy as np

# Data input (X, Y)
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 3, 5, 7, 11])

# Garis regresi linear
m, b = np.polyfit(X, Y, 1)
regression_line = m * X + b

# Halaman 1: Input Data (Table)
# Disarankan untuk menyajikan ini dalam bentuk tabel di buku, tidak menggunakan grafik.

# Halaman 2: Formula f(x)
# Disarankan untuk menyajikan ini dalam bentuk teks dan formula matematika di buku.

# Halaman 3: Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Data Points')
plt.title('Scatter Plot of Data')
plt.xlabel('X (Input)')
plt.ylabel('Y (Output)')
plt.legend()
plt.grid(True)
plt.show()

# Halaman 4: Proses Menentukan Garis Regresi
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, regression_line, color='red', label='Regression Line')
plt.title('Scatter Plot with Regression Line')
plt.xlabel('X (Input)')
plt.ylabel('Y (Output)')
plt.legend()
plt.grid(True)
plt.show()

# Halaman 5: Garis Regresi
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, regression_line, color='red', label='Regression Line')
plt.title('Linear Regression Line')
plt.xlabel('X (Input)')
plt.ylabel('Y (Output)')
plt.text(1, 10, f'Y = {round(m, 2)}X + {round(b, 2)}', fontsize=12, color='red')
plt.legend()
plt.grid(True)
plt.show()

# Halaman 6: Penjelasan Graf Cartesian
plt.figure(figsize=(10, 6))
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.scatter(X, Y, color='blue', label='Data Points')
plt.title('Cartesian Graph Explanation')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Halaman 7: Contoh Nyata (Prediksi Harga Rumah berdasarkan Luas Tanah)
X_example = np.array([1000, 1500, 2000, 2500, 3000])
Y_example = np.array([200000, 250000, 300000, 350000, 400000])

# Garis regresi linear untuk contoh nyata
m_example, b_example = np.polyfit(X_example, Y_example, 1)
regression_line_example = m_example * X_example + b_example

plt.figure(figsize=(10, 6))
plt.scatter(X_example, Y_example, color='blue', label='Data Points')
plt.plot(X_example, regression_line_example, color='red', label='Regression Line')
plt.title('House Price Prediction based on Land Size')
plt.xlabel('Land Size (sq ft)')
plt.ylabel('House Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```
