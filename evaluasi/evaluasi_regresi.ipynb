```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Simulated ground truth and predicted values
# Nilai aktual (ground truth) yang sebenarnya
# dan nilai prediksi dari model (predicted_values)
ground_truth = np.array([2.5, 3.7, 4.2, 5.0, 6.1])
predicted_values = np.array([2.2, 3.5, 4.0, 4.8, 5.8])

# Menghitung Mean Squared Error (MSE)
# MSE mengukur rata-rata kuadrat selisih antara nilai aktual dan prediksi
mse = mean_squared_error(ground_truth, predicted_values)

# Menghitung Root Mean Squared Error (RMSE)
# RMSE adalah akar kuadrat dari MSE, memberikan nilai error dalam satuan yang sama dengan data asli
rmse = np.sqrt(mse)

# Menghitung Mean Absolute Error (MAE)
# MAE mengukur rata-rata selisih absolut antara nilai aktual dan nilai prediksi
mae = mean_absolute_error(ground_truth, predicted_values)

# Mengonversi MAE ke dalam persentase dari rata-rata nilai aktual (ground_truth)
# Ini memberi gambaran lebih jelas tentang seberapa besar kesalahan rata-rata dibandingkan dengan data asli
mae_percentage = (mae / np.mean(ground_truth)) * 100

# Menghitung Mean Absolute Percentage Error (MAPE)
# MAPE mengukur kesalahan prediksi dalam bentuk persentase terhadap nilai aktual
mape = np.mean(np.abs((ground_truth - predicted_values) / ground_truth)) * 100

# Menampilkan hasil-hasil evaluasi
# MSE, RMSE, MAE dalam satuan nilai absolut dan MAE dalam persentase terhadap rata-rata ground_truth
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae} ({mae_percentage:.2f}% of average ground truth)")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")

# Visualisasi perbandingan antara nilai aktual (ground_truth) dan nilai prediksi
plt.figure(figsize=(10, 5))

# Plot nilai aktual dan prediksi
plt.plot(ground_truth, label='Ground Truth', marker='o')  # Ground truth
plt.plot(predicted_values, label='Predicted Values', marker='x')  # Predicted values

# Menambahkan area berwarna abu-abu untuk menunjukkan perbedaan antara nilai aktual dan prediksi
plt.fill_between(range(len(ground_truth)), ground_truth, predicted_values, color='gray', alpha=0.2)

# Memberikan judul, label, dan legenda pada grafik
plt.title('Ground Truth vs Predicted Values')
plt.xlabel('Data Points')
plt.ylabel('Values')
plt.legend()

# Menampilkan grafik
plt.show()
```
