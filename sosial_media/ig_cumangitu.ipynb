```python
import pandas as pd

# LOad data csv 
file_path = '../data-ig.csv'
data = pd.read_csv(file_path)

# Menampilkan 5 baris teratas untuk tau struktur row dari data kita 
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation matrix
correlation_matrix = data[['Impresi', 'Jangkauan', 'Frekuensi Dibagikan', 'Mengikuti', 'Diputar', 'Suka', 'Komentar', 'Frekuensi Disimpan']].corr()

# Display the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Engagement Metrics')
plt.show()
```

```python
# Calculating the mean of engagement metrics to understand what types of content are most engaging
mean_metrics = data[['Durasi (Detik)', 'Impresi', 'Jangkauan', 'Frekuensi Dibagikan', 'Mengikuti', 'Diputar', 'Suka', 'Komentar', 'Frekuensi Disimpan']].mean()

# Analyzing the top 10 performing posts based on the 'Suka' (Likes) metric
top_10_posts = data.nlargest(10, 'Suka')

```
